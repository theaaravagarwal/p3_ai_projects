# CNN Filter Demo\nThis notebook applies a 3x3 convolution kernel to a tiny binary image.

```python
import numpy as np\nfrom scipy.signal import convolve2d\nfrom matplotlib import pyplot as plt\n\n# Image of an 'X'\nimage = np.array([\n    [1, 0, 0, 0, 1],\n    [0, 1, 0, 1, 0],\n    [0, 0, 1, 0, 0],\n    [0, 1, 0, 1, 0],\n    [1, 0, 0, 0, 1],\n], dtype=np.uint8)\n\nkernel = np.array([\n    [1, -1, -1],\n    [-1, 1, -1],\n    [-1, -1, 1],\n])\n\n# Convolution is the key operation behind CNN filters.\nresult = convolve2d(image, kernel, mode='same', boundary='fill', fillvalue=0)\nprint('Result of convolution:')\nprint(result)\n\nfig, axes = plt.subplots(1, 2, figsize=(8, 4))\naxes[0].imshow(image, cmap='gray', vmin=image.min(), vmax=image.max())\naxes[0].set_title('Input Image')\naxes[0].axis('off')\n\naxes[1].imshow(result, cmap='gray')\naxes[1].set_title('Convolution Output')\naxes[1].axis('off')\n\nplt.tight_layout()\nplt.show()
```
