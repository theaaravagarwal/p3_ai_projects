```python
import numpy as np
from scipy.signal import convolve2d
from matplotlib import pyplot as plt

# Image of a "T"
image = np.array([[1, 0, 0, 0, 1],
                  [0, 1, 0, 1, 0],
                  [0, 0, 1, 0, 0],
                  [0, 1, 0, 1, 0],
                  [1, 0, 0, 0, 1],
                  ], np.uint8)

plt.imshow(image, cmap='gray', vmin=image.min(), vmax=image.max())
plt.show()

kernel = np.array([[1, -1, -1],
                   [-1, 1, -1],
                   [-1, -1, 1]])

# Write your code below perform the convolution and applying the kernel
result = convolve2d(image, kernel, mode="same")
print("Result of convolution: ")

plt.imshow(result, cmap='gray', vmin=result.min(), vmax=result.max())
plt.show()
```
